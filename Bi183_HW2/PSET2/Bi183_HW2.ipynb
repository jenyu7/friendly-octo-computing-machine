```python
import numpy as np

# given two bases, determines score
def score(a, b): 
    if a == b:
        return match
    else:
        return mismatch

def needleman_wunsch(seq1, seq2):   
    # score matrix of dimension m+1 x n+1
    n = len(seq1)  
    m = len(seq2)
    
    # initialize matrix to store scores
    scores = np.zeros((m+1, n+1))
   
    # fill out first column and row with gaps
    # prevents needing to check for edge cases later in loop
    for i in range(0, m + 1):
        scores[i, 0] = gap * i
    for j in range(0, n + 1):
        scores[0, j] = gap * j
    print(scores)
    # populate the matrix with scores
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            match = scores[i - 1, j - 1] + score(seq1[j-1], seq2[i-1])
            delete = scores[i - 1, j] + gap
            insert = scores[i, j - 1] + gap
            # store the maximum score 
            scores[i, j] = max(match, delete, insert)
            print(scores)
    
    # after score matrix has been calculated, backtrack to find best alignment
    # add to a string as backtracking occurs (note that this sequence will be backwards)
    
    # start backtracking from bottom right cell
    seq1_align = ""
    seq2_align = ""
    i = m
    j = n
    # loop ends when touching the top or left edge
    while i > 0 and j > 0: 
        current = scores[i, j]
        # obtain scores from neighboring cells
        diagonal = scores[i-1, j-1]
        up = scores[i, j-1]
        left = scores[i-1, j]
        
        # find which cell produced the max score
        # either match or mismatch
        if current == diagonal + score(seq1[j-1], seq2[i-1]):
            seq1_align += seq1[j-1]
            seq2_align += seq2[i-1]
            # move diagonally
            i -= 1
            j -= 1
        # gaps: move up or left
        elif current == up + gap:
            seq1_align += seq1[j-1]
            seq2_align += '-'
            # move up
            j -= 1
        elif current == left + gap:
            seq1_align += '-'
            seq2_align += seq2[i-1]
            # move left
            i -= 1

    # add in gaps to corresponding sequence until 0, 0 is reached
    while j > 0:
        seq1_align += seq1[j-1]
        seq2_align += '-'
        j -= 1
    while i > 0:
        seq1_align += '-'
        seq2_align += seq2[i-1]
        i -= 1
    
    # backtracking created a backwards sequence -- reverse it
    seq1_align = seq1_align[::-1]
    seq2_align = seq2_align[::-1]

    return seq1_align, seq2_align, scores[m][n]

seq1 = "GATTACA"
seq2 = "AATGGACA"
match = 1
mismatch = -1
gap = -2
print(needleman_wunsch(seq1, seq2))
# match > gaps or mismatches
# finds and lists all possible alignments by varying match, mismatch and gap
'''
alignments1 = []
alignments2 = []
for i in range(-10, 10): 
    for j in range(-10, 10):
        for k in range(-10, 10):
            match = i
            mismatch = j
            gap = k
            a, b, sc = needleman_wunsch(seq1, seq2)
            if a not in alignments1:
                alignments1.append(a)
            if b not in alignments2:
                alignments2.append(b)
'''
#print(alignments1)
#print(alignments2)
#print(needleman_wunsch(seq1, seq2))
```

## Problem 6 
**Extended to 4 observed states**

```python
import copy

class HMM:
    def __init__(self, states, transition, emission, init):
        self.state_names = copy.copy(states)
        self.n_states = len(states)
        self.A = transition.copy()
        self.B = emission.copy()
        self.n_emissions = self.B.shape[1]
        self.init = init

    def generate(self, length):
        state = self.init
        states = []
        ret = []
        for i in range(1, length + 1):
            state = np.random.choice(range(self.n_states), p=self.A[state])
            states.append(state)
            ret.append(
                np.random.choice(range(self.n_emissions), p=self.B[state]))
        print('Generating by states:', ''.join(self.state_names[i] for i in states))
        ret = ''.join([str(i) for i in ret])
        return ret
    

    def viterbi(self, seq):
        # := max-product
        seq_arr = np.array([int(i) for i in seq])
        T = len(seq_arr)
        T1 = np.zeros((self.n_states, T + 1))
        T1[self.init, 0] = 1
        T2 = np.zeros((self.n_states, T + 1), dtype='int')
        states = np.zeros(T + 1, dtype='int')
        for t in range(1, T + 1):
            for j in range(self.n_states):
                T1[j, t] = np.max(T1[:, t - 1] * self.A[:, j])
                T1[j, t] *= self.B[j, seq_arr[t - 1]]
                T2[j, t] = np.argmax(T1[:, t - 1] * self.A[:, j])
        states[T] = np.argmax(T1[:, T])
        for t in range(T, 1, -1):
            states[t - 1] = T2[states[t], t - 1]
        return ''.join([self.state_names[s] for s in states[1:]])
    
np.random.seed(1236)
states = ['A', 'B', 'C', 'D']

print("1.1 Generation\n")
transition = np.array([
    [0.8, 0.1, 0.0, 0.1],
    [0.1, 0.5, 0.2, 0.2],
    [0.1, 0.0, 0.6, 0.3],
    [0.2, 0.3, 0.1, 0.4]
])
emission = np.array([
    [0.9, 0.1],
    [0.5, 0.5],
    [0.1, 0.9], 
    [0.2, 0.8]
])
init = 0
hmm = HMM(states, transition, emission, init)
seqs = []
for seq_len in [100, 1000, 10000]:
    seq = hmm.generate(seq_len)
    seqs.append(seq)
    print("Inferred optimal state series:")
    print(hmm.viterbi(seq))
    # NOTE: To run chains with various length, REMOVE this break
    break
```
